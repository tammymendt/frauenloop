```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./climate-change-earth-surface-temperature-data/GlobalLandTemperaturesByCountry.csv",
                parse_dates=['dt'])
```

```python

df.head()
```

```python
df.query("Country == 'Germany'")[['dt', 'AverageTemperature']].plot(x='dt', rot=90)
```

```python
df.query("Country == 'Germany'").head()
```

```python
jan_avg_temps_df = df.query("Country == 'Germany'")[df['dt'].dt.month == 1].groupby(df['dt'].dt.year).agg({'AverageTemperature': 'mean'}).reindex()
```

```python
jan_avg_temps_df.plot()
```
